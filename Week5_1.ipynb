```python
class Node:
    def __init__(self,item, pointer=None):
        self.item = item
        self.next = pointer

class LinkedList:

    def __init__(self):
        self._head = None
        self._tail = None
        self._size = 0

    def addFirst(self, item):
        node = Node(item)
        node.next = self._head
        self._head = node
        if self._size == 0:
            self._tail = node
        self._size += 1

    def removeFirst(self):
        if self._size == 0 :
            raise IndexError("error")
        self._head = self._head.next
        self._size -= 1
        if self._head == None:
            self._tail = None
            
    def addLast_noTail(self,item):
        node = Node(item)
        walk = self._head
        while walk !=None and walk._next != None:
            walk = walk._next
        walk._next = node
        if self._size == 0:
            self._head = node
        self._size += 1
     
    def addLast_withTail(self,item):
        node = Node(item)        
        if self._size == 0:
            self._head = node
            self._tail = self._head
        else:
            self._tail.next = node
        self._size += 1
        
    def removeLast(self):
        if self.is_empty():
            raise IndexError('Error')
        if self._head == self._tail:
            # only one element
            self._head = None
            self._tail = None
        else:
            walk = self._head
            while walk.next is not self._tail:
                walk = walk.next
            walk.next = None
            self._tail = walk
        self._size -= 1

```
